# Reported Measles Cases and Outbreaks at the County Level in the U.S.

```python
# remove intro and other non-state sheets
# ref: https://www.pythonexcel.com/delete-sheet.php
import openpyxl
file_name = r".\US Measles Cases and Outbreaks Summary (2019).xlsx"
workbook=openpyxl.load_workbook(file_name)

pre_list = workbook.sheetnames[0:3]
for i in pre_list:
    del workbook[i]
    print (i + " removed.")
    
workbook.save(file_name)
```

```python
# ref: https://pbpython.com/pandas-excel-tabs.html
import pandas as pd
import numpy as np
#file_name = r".\US Measles Cases and Outbreaks Summary (2019).xlsx"
df = pd.concat(pd.read_excel(file_name, sheet_name=None), ignore_index=True, sort=False)
print("Total number of counties:",len(df))
```

```python
df_cases = df[df['# of Cases'].notnull()]
df_cases.loc[df_cases['int_o'].notnull() | df_cases['int_no'].notnull(), 'intl_case'] = 1
print(len(df_cases))
df_cases.head()
```

```python
result = df_cases
output_csv = './reported_cases.csv'
result.to_csv(output_csv, index=False, encoding='utf-8')
```
