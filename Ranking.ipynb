# Risk Ranking by Year

```python
import pandas as pd
```

```python
# load data
folder = r'C:\Users\Ensheng\Desktop\mapping\measles_county_year\\'
in_table = folder + 'ranking_input.xlsx'
raw_df = pd.read_excel(in_table)
raw_df.head(5)
```

```python
# select important fields
df = raw_df[['OBJECTID','FIPS','Risk','Year']]
df = df.rename(index=str, columns={"OBJECTID": "RankObjID"})
df.head(5)
```

```python
# sort by risk and group by year (AMAZING!)
df['Risk_Rank'] = df.groupby('Year')['Risk'].rank(ascending=False,method='dense')
```

```python
df.sort_values(by=['Risk'], ascending=True).head(15)
```

```python
# output as a csv table
result = df
output_csv = folder + 'ranking_output.csv'
result.to_csv(output_csv,index = False,encoding='utf-8')
```
