#### Objective:
Create a time slider to describe counts of measleas reports for all countries in the world from 2011 to 2019 yearly.

#### Update Info:
* 04/17/2019: first version based on MIA_MRE.

```python
import pandas as pd
```

```python
in_table = r'C:\Users\Ensheng\Desktop\mapping\world_year\infectedCountries_till2019.csv'
df = pd.read_csv(in_table)
len(df)
```

```python
df['TimeID'] = df['ISO3'] + df['Year'].map(str)
df.head(15)
```

```python
in_table = r'C:\Users\Ensheng\Desktop\mapping\measles_world_year\world_measles_year.csv'
iso3_df = pd.read_csv(in_table)
iso3_df['TimeID'] = iso3_df['ISO_3DIGIT'] + iso3_df['Year'].map(str)
iso3_df.head(15)
```

```python
len(iso3_df)
```

```python
# merge iso3_df and df
result = pd.merge(iso3_df, df, how='outer', on='TimeID')
result['GISID'] = result['OBJECTID']
result = result[['GISID','ISO_3DIGIT','Year_x','Total']]
result[130:140]
```

```python
# output as a csv table
output_csv = r'C:\Users\Ensheng\Desktop\mapping\world_year\world_year.csv'
result.to_csv(output_csv,index = False,encoding='utf-8')
```
