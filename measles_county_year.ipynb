# measles_county_year

```python
import pandas as pd
```

```python
folder = r'C:\Users\Ensheng\Desktop\mapping\measles_county_year\\'
in_table = folder + 'ModelInputOutputAll 4_23.csv'
raw_df = pd.read_csv(in_table)
raw_df.head(5)
```

```python
list(raw_df)
```

```python
common_list = ['County Name','State','FIPS'] # Be cautious: space before the label
year_list = ['2011','2012','2013','2014','2015','2016','2017','2018','2019']
#year_list = [2011,2012,2013,2014,2015,2016,2017,2018,2019]
```

```python
# melt and sort table
df = pd.melt(raw_df, id_vars=common_list,value_vars=year_list)
# rename add month
df.loc[:,('Year')] = df['variable']
df.loc[:,('Measles')] = df['value']
df = df[['County Name','State','FIPS', 'Year', 'Measles']]
df.head(5)
```

```python
# merge year and month
df['TimeID'] = df['FIPS'].map(str) + df['Year'].map(str)
df = df[['FIPS', 'State', 'Year', 'Measles', 'TimeID']]
df.head(5)
```

### Data from ArcMap

```python
in_table = folder + 'OBJECTID_GEOID.csv'
iso3_df = pd.read_csv(in_table)
iso3_df = iso3_df.sort_values(by=['OBJECTID'])
iso3_df.head(5)
```

```python
import numpy as np
x1 = ['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
x = np.repeat(x1,3220)
iso3_df['Year'] = x
iso3_df.head(5)
```

```python
iso3_df['TimeID'] = iso3_df['GEOID'].map(str) + iso3_df['Year'].map(str)
```

```python
iso3_df.head(5)
```

```python
# merge iso3_df and df
result = pd.merge(iso3_df, df, how='outer', on='TimeID')
result['GISID'] = result['OBJECTID']
result['Risk'] = result['Measles']
result['Year'] = result['Year_x']
result = result[['GISID','GEOID','FIPS','Risk','TimeID','Year','State']]
result[130:140]
```

```python
# output as a csv table
output_csv = folder + 'output.csv'
result.to_csv(output_csv,index = False,encoding='utf-8')
```
